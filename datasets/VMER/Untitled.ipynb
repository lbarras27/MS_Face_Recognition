```python
from retinaface import RetinaFace
import cv2
import os, sys
```

```python
img_path = "test/n000001/0001_01.jpg"
faces = RetinaFace.detect_faces(img_path)
```

```python
img = cv2.imread(img_path)
```

```python
facial_area = faces["face_1"]["facial_area"]
landmarks = faces["face_1"]["landmarks"]
cv2.rectangle(img, (facial_area[2], facial_area[3]),(facial_area[0], facial_area[1]), (255, 255, 255), 1)
cv2.circle(img, tuple([int(e) for e in landmarks["left_eye"]]), 1, (0, 0, 255), -1)
cv2.circle(img, tuple([int(e) for e in landmarks["right_eye"]]), 1, (0, 0, 255), -1)
cv2.circle(img, tuple([int(e) for e in landmarks["nose"]]), 1, (0, 0, 255), -1)
cv2.circle(img, tuple([int(e) for e in landmarks["mouth_left"]]), 1, (0, 0, 255), -1)
cv2.circle(img, tuple([int(e) for e in landmarks["mouth_right"]]), 1, (0, 0, 255), -1)
```

```python
cv2.imwrite("img_test.jpg", img)
```

```python
all_img_names = []
for identity in os.listdir("test"):
    for img_name in os.listdir(os.path.join("test", identity)):
        all_img_names.append(identity+"/"+img_name)        
```

```python
len(all_img_names)
```

```python
all_landmarks = {}
for i, img_path in enumerate(all_img_names):
    faces = RetinaFace.detect_faces("test/" + img_path)
    if type(faces) == dict:
        landmarks = faces["face_1"]["landmarks"]
    else:
        landmarks = []
        
    all_landmarks[img_path] = landmarks
    if i % 100 == 0:
        print(i)
```
