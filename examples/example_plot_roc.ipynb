```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
results = np.load("all_fpr_tpr_auc.npy", allow_pickle='True').item()
```

```python
### Plot ROC curve for Webface

target = "Webface-OCC"
webface = results["webface"]
labels = ["Without occlusion", "Glasses occlusion", "Surgical mask occlusion"]

for occ_type, label in zip(webface.keys(), labels):
    webface_occ = webface[occ_type]
    plt.plot(webface_occ["fpr"],
                webface_occ["tpr"],
                lw=1,
                label=('%s (AUC = %0.4f %%)' %
                    (label, webface_occ["auc"] * 100)))

plt.xlim([10 ** -6, 1])
plt.ylim([0.4, 1])
plt.xscale('log')
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC on {}'.format(target))
```
